```python
import pandas as pd
```

# bigqueryに読み込みをした際にカラム名が日本語だと取り込めないため英語に変更


```python
info_data = pd.read_csv('../data/整形後データ/bike_info.csv',index_col=0)
model = pd.read_csv('../data/整形後データ/modelnumber.csv',index_col=0)
tenpo = pd.read_csv('../data/整形後データ/tenpo_data.csv',index_col=0)
```

```python
info_data
```

```python
info_data.rename(columns={'価格':'cost',
                          '型式番号':'model_number',
                          'モデル年':'model_year',
                          '色':'color',
                          '走行距離':'mileage',
                          '保証期間':'guarantee_period',
                          '店舗id':'store_id'},inplace=True)
```

```python
model.rename(columns={'型式番号':'model_number',
                      'メーカー':'brand',
                      '車両名':'name',
                      'タイプ':'type',
                      '排気量':'displacement'},inplace=True)
```

```python
tenpo_id = tenpo['店舗id']
tenpo.drop(columns='店舗id',inplace=True)
tenpo.insert(0,'store_id',tenpo_id) 
tenpo.rename(columns={'店舗名':'name',
                      '郵便番号':'postal_code',
                      '住所':'address',
                      '電話番号':'phone'},inplace=True)
```

```python
info_data.to_csv('../data/アップロード用/bike_info.csv')
model.to_csv('../data/アップロード用/model_data.csv')
tenpo.to_csv('../data/アップロード用/tenpo_data.csv')
```
