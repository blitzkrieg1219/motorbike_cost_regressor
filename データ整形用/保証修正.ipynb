保証は年と月が混ざっおり、アップロードデータがおかしいため修正

```python
import pandas as pd
```

```python
df_origin = pd.read_csv('../data/元データ/df_info_20231115.csv',index_col=0)
df_target = pd.read_csv('../data/アップロード用/bike_info.csv',index_col=0)
```

```python
df_origin
```

```python
#元データから真の値を作り直す(年に統一)
#元データを一度strに変換してから実行する(intが含まれていると.strが使えないため)
df_origin['保証期間'] = df_origin['保証期間'].astype(str)
def grantee_period_replace(text):
    if text[1:] == 'ヶ月':
        text = int(text[0])/12
    return text
```

```python
df_origin['保証期間']=df_origin['保証期間'].apply(grantee_period_replace)
```

```python
#保証期間が'-'のものは0を入れておく
df_origin['保証期間']=df_origin['保証期間'].replace('-',0)
df_origin['保証期間'] = df_origin['保証期間'].astype(float)
```

```python
df_target['guarantee_period'] = df_origin['保証期間'].values
```

```python
df_target.to_csv('../data/アップロード用/bike_info.csv')
```
