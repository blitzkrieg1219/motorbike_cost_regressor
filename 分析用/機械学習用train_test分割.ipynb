bike_infoだけで良い

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#data読み込み
data = pd.read_csv('../data/アップロード用/bike_info.csv',index_col=0)
data.dropna(inplace=True)
```

```python
test_data = data.sample(int(len(data)*0.2))
train_data = data.drop(test_data.index)
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
#偏りがないかを確認
sns.pairplot(train_data.drop(columns=['id','model_number','color']))
plt.show()
```

```python
train_data.to_csv('../data/機械学習用/train_data.csv')
test_data.to_csv('../data/機械学習用/test_data.csv')
```
